```python
import pandas as pd
import requests
```

```python
source = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
table = requests.get(source)

df1 = pd.read_html(table.content, header=0)[0]
df2 = df1[df1.Borough != 'Not assigned']
df2.head()
```
